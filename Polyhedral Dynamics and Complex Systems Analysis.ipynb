## Abstract
Polyhedral dynamics, more commonly known as [Q-analysis](https://en.wikipedia.org/wiki/Q-analysis) is a mathematical framework for describing and analyzing the structural properties of relations between sets. It was originally developed to study the structure and dynamics of social systems ([Atkin, 1972](https://www.sciencedirect.com/science/article/pii/S0020737372800294)), but because of its roots in algebraic topology, q-analysis can be generalized to deal with a wide range of complex systems. It has been applied to the study of chess ([Atkin & Witten, 1977](http://www.sciencedirect.com/science/article/pii/S0020737375800356)), urban transportation systems ([Johnson, 1981](http://journals.sagepub.com/doi/abs/10.1068/b080141), Johnson, 1984), topic analysis ([Jacobson & Yan, 1998](https://link.springer.com/article/10.1023%2FA%3A1004255229611?LI=true), [Rodrigues, 2013](http://oro.open.ac.uk/42634/)), multi-criteria decision making ([Heissl et. al., 1985](http://www.sciencedirect.com/science/article/pii/0096300385900219), [Chin et. al., 1991](http://www.sciencedirect.com/science/article/pii/009630039190019J), [Duckstein & Nobe, 1997](http://www.sciencedirect.com/science/article/pii/S0377221797003081), [Bhattacharyya & Chakraborty, 2015](http://growingscience.com/dsl/Vol4/dsl_2014_33.pdf)),, multiplexity in entrepreneurship networks ([Bliemel et. al.](https://s3.amazonaws.com/academia.edu.documents/35771791/2014_Multiplexity.pdf?AWSAccessKeyId=AKIAIWOWYYGZ2Y53UL3A&Expires=1513826618&Signature=1prc261i%2BsslmGqYDcGSg6RoSXU%3D&response-content-disposition=inline%3B%20filename%3DAn_Integrated_Approach_to_Studying_Multi.pdf)), and developing expert systems for medical diagnostics. The notebook covers several key concepts from q-connectivity, q-transmission, eccentricity, and traffic. The concepts and methods provide some of the initial building blocks for developing 'intelligent' systems, and serve as the basic building blocks for constructing multi-level complex systems that can be used in multi-criteria decision-making and analysis.

## Introduction
The foundations for graph theory (and later topology) were first developed by the mathematician, Leonard Euler in 1736 when he tackled the famous Seven Bridges of Königsberg problem. Euler's task was to identify a route where one could walk through the city while crossing all seven bridges once, and only once. Although a problem of little consequence, the Seven Bridges presented an interesting challenge, as the mathematics of the day were ill-suited to solving the problem. Euler realized that the problem was not a counting problem or workable using standard Euclidian geometry.

### Figure 1. Seven Bridges of Königsberg

![Figure 1. Seven Bridges of Königsberg](https://plus.maths.org/content/sites/plus.maths.org/files/blog/112013/bridges.jpg)

Instead, Euler devised a way to define the 'geometry of position' by reducing the problem system to its invariant properties, as a set of vertices (land masses) and edges (bridges) resulting in a graph that could be easily inspected. Euler proved that it was impossible for a person to walk across all seven bridges in single circuit of the city. Euler's solution laid the groundwork for an entirely new branch of mathematics, known as algebraic or combinatorial topology.

Although topology has been around in some form for nearly 300 years, the field has exploded in the last 60 years. Mathematician and researchers both have utilized topological methods to tackle a number of complex problems, leading to major breakthroughs in physics, geography, computer science, neuro-sciences, urban planning and design, and even artificial intelligence ([Carlsson, 2009](http://www.ams.org/journals/bull/2009-46-02/S0273-0979-09-01249-X/S0273-0979-09-01249-X.pdf), [Johnson 2014](http://www.worldscientific.com/worldscibooks/10.1142/p533)). Topology has also influenced scientific discoveries related to the function and structure of complex systems, swarm intelligence, and more.

Polyhedral dynamics represents one of the major innovations of topology that arose during the last 60 years of development. Based primarily on the work of Ron Atkin ([Atkin, 1972](https://www.sciencedirect.com/science/article/pii/S0020737372800294), [Atkin & Casti 1977](http://pure.iiasa.ac.at/709/)), polyhedral dynamics offers a mathematical system capable of describing and analyzing the structural and dynamic properties of relations between sets of simplicies and vertices. The combinatorial connectives linking simplicies and vertices form the simplicial complex as a representation of a multi-dimensional system upon which the dynamical behavior of the system can be directly inspected.

### Figure 2. Example - Simplicial Complex

![http://doi.ieeecomputersociety.org/10.1109/TVCG.2012.248](http://doi.ieeecomputersociety.org/cms/Computer.org/dl/trans/tg/2012/12/figures/ttg20121223823.gif)

More recent developments in hypernetwork theory ([Johnson 2006](http://oro.open.ac.uk/4628/); [2014](http://www.worldscientific.com/worldscibooks/10.1142/p533); [2016](https://link.springer.com/article/10.1140/epjst/e2016-02653-4)) greatly expand the application space of q-analysis by providing a formalism for describing n-ary relations, and ordered set. The expansion beyond the binary relation opens up new avenues for studying complex systems, including multi-level systems of systems where space and time (as position in time) are but two dimensions in an n-dimensional topological space. These additions also present some interesting opportunities for exploring topological approaches to artificial intelligence, particularly, knowledge representation such ontological engineering, classification and complex multi-attribute decision-making, and optimization.

However, for the purposes of this notebook, we will simply review some of the basic foundations of the algebraic system, as this will provide a reusable reference for the development and experimentation of polyhedral systems to study language, innovation, markets, and eventually intelligence.

## Foundations
The foundation of polyhedral dynamics rests on the principles of algebraic topology and simplicial homology theory (Casti) to extract the topological invariants of a simplicial complex representing a given system under investigation. Constructing the system complex is based on defining the relation between two well-defined sets given a set of values describing the character of the relation. For example, the relation defined between some set A and another set B produce observable geometric properties e.g. simplicies, vertices, edges and faces.

![\lambda \subseteq A \cdot  B](https://3.bp.blogspot.com/-Gpnv-YekDt4/WW1WGy5X3mI/AAAAAAAAEl4/ooxYG0h1k6cbF-sUT6CG1cGvjDjgHs8qgCPcBGAYYCw/s320/Screen%2BShot%2B2017-07-17%2Bat%2B5.25.43%2BPM.png)

The simplicial complex reveals the structural character of data based on a relational mapping between A and B, for a given threshold or slicing parameter \theta that determine the conditions under which elements in the system are retained or dropped. Given this slicing parameter, the relational mapping produces in an incidence matrix that highlight the connectivities between set A (the collection of simplicies), and set B (the collection of vertices) that describe each individual simplex.

In a computational framework, constructing the incidence matrix relies on defining three specific elements:

1. A data matrix, 
2. A label set (A) representing the rows in the matrix, and 
3. A label set (B) representing the columns in the matrix. 

Two or more additional sets can be included to compute weighted relations and dynamical properties over the complex. For sake of simplicity, we will focus on the two sets and the data matrix which are defined below as two flat Python lists and one 2-d list. Of course, if requiring greater computational efficiency, these data could be defined as  [numpy arrays](https://docs.scipy.org/doc/numpy-1.13.0/user/quickstart.html).

### Figure 3. Defined Sets & Data Matrix

```python
A = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']

B = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']

matrix = [[0, 1, 7, 0, 1, 6, 3, 2], 
          [5, 4, 7, 0, 0, 1, 4, 1],
          [1, 1, 1, 6, 6, 4, 3, 5],
          [1, 3, 7, 7, 3, 5, 5, 4],
          [1, 6, 6, 5, 4, 0, 0, 0],
          [1, 1, 1, 2, 4, 0, 0, 7],
          [0, 2, 2, 5, 0, 0, 0, 6],
          [0, 3, 3, 4, 4, 0, 0, 0],
          [0, 0, 7, 1, 0, 0, 1, 1],
          [2, 7, 2, 2, 0, 0, 1, 6],
          [1, 1, 1, 0, 4, 0, 0, 0],
          [3, 6, 6, 0, 1, 0, 0, 1]]
```

In this case, the threshold parameter mentioned earlier serves to define which elements are retained and which are ignored. In this example, we might define a threshold value to equal 5 in which case all elements that are equal to or greater than 5 are given a value of 1, and all other values are given a value of 0.

The Python function below provides a simple implementation of an incidence matrix constructor. ### Computing the Conjugate Complex and the Q-Matrix
In addition to defining the incidence matrix based on the simplicies as rows, it is also possible to consider the vertices as simplicies and the simplicies as vertices by taking the conjugate complex of the original incidence matrix. This is simply computed by taking the transpose of the original matrix. This is computed using the computeConjugate() function, which takes the incidence matrix and the two label sets as input parameters. 

### Figure 7. Conjugate Complex

```python
Conjugate = computeConjugate(C, A, B)
Conjugate[1]
```

The next step in the process is to compute the shared-face matrix between simplicies that reveal the number of vertices representing each simplex, and the relation between each simplex as defined by the number of shared vertices. This accomplished by multiplying the original incidence matrix by its transpose, depicted below in the computeQFace() function.

### Figure 8. Shared Face Matrix for the Simplex Set A

```python
faces = computeQFace(Conjugate[0], Conjugate[1], A, B)
faces
```

The shared-face matrix table reveals the number of vertices shared between each simplex. The q-dimension for the shared-face matrix is computed by subtracting 1 from the value of each ij location in the matrix. The resulting computation produces the q-dimensions, where a simplex with 2 verticies is represented as a 1-dimesnional simplex, a 3 vertex simplex produces a 2-dimension simplex represented as a triangle, whereas a 3-dimensional simplex forms a 4 tetrahedron.

### Figure 9. Shared Face Q-Matrix

```python
QMatrix = computeQMatrix(faces, A)
QMatrix
```

The next step in the procedure involves extracting the q-connected components that form the simplicial complex representing the entire system. At different levels of q, the structure of these components can and do change, allowing us to explore the system interactions at different dimensional levels. This is important feature of q-analysis in that by focusing on the connectivity of simplicies and the degree to which they are embedded in the system (simplicial complex) can be interpreted to identify control points, or critical failure points depending on the focus of analysis.

The q_components() function below is used to extract the components (complexes) that describe the system at a given q-level. It is important to note that at this time that our analysis is focused on first identifying q-near and q-connected simplicies, as these features are central parameters to computing 'traffic' patterns over the complex. But before we proceed it is necessary to define and differentiate these two terms in the context of our analysis. Simplicies that are q-near are defined by the fact that the simplicies share at least one vertex. On the other hand, a q-connected simplicies may not share a vertex, but instead can be connected to another simplex in which the two original simplicies share at least one vertex. So, in this way, a simplicies might be q-connected, but not q-near.

The emphasis on q-connected components will become clearly when computing dynamic traffic patterns over the complex as the q-connected components provide a window into available pathways for spread of a particular event. This spreading phenomenon is called q-transmission and is a central feature in the q-analysis system. For instance, in our hypothetical example of academic units at a university, the measure of q-connectivity could be used to measure the level of integration between skills and units to accomplish a given task. By removing a given person from a unit, we might find that an obstruction occurs where a particular person and skill-set severed to link the unit to other units in the university. Q-transmission provides a measure of the impact on the system as members of each unit attempt to connect and collaborate. It might also be used to assess the likely collaborators at person level and at the unit level. Again, what is important to remember here is that the q-connectivity provides one of the initial measures for computing the q-transmission of events or behavior over the system.

Johnson provides a detailed description of q-transmission in his article, "q-Transmission in simplicial complexes"([1982](http://www.sciencedirect.com/science/article/pii/S0020737382800461) and is certainly worth reviewing to better understand the importance of q-transmission patterns in the simplicial complex. Yet first, we must identify the connected components to assess the level of connectivity that ultimately determine transmission behavior.

## Mapping Traffic
The description and analysis of dynamics of a simplicial complex in the polyhedral system is addressed through the concept of 'traffic.' While most general concepts of traffic relate to movement of objects, traffic in this context simply denotes the presence of something existing over the 'backcloth' or complex. The complex provides the structure for determining the potential flow of entities. The road system on which cars travel represents the complex or backcloth. As a rule-of-thumb, vehicles do not deviate from the built roadway, and instead follow the networked pathway to reach a given destination. If places of interest are fragmented, then it may be, structurally speaking inaccessible, meaning the gaps (q-holes) serve as obstructions to the existence of traffic, limiting or eliminating accessibility to a given location in the structure.

The mapping of traffic is typically an easy mapping of simplicies into a set of integers representing the value for a given traffic at a specified dimension. This allows us to extend the original matrix by constructing a new set that can be evaluated. Additionally, the mapping of integers could be multiplied by a given set of weights based on some theoretical or preference ordering. This enables q-analysis to serve as a decision-aide in multi-criterion decision-making (see: Duckstein and Nobe, 1997).

For our example here, we will start by creating two new sets that represent pre and post values. These two sets of integers will be used to measure change in traffic over the complex.

```python
pre = [10, 12, 3, 10, 5, 19, 17, 0, 0, 9, 8, 1]
post = [7, 12, 3, 20, 8, 12, 17, 2, 1, 5, 6, 5]

traffic = map_traffic(A, pre, post) #needs work and needs to be linked to q-transmission
traffic
```

The results of this function can be interpreted in a number of ways, but most commonly we would think of the traffic value in the third column as an attractor or t-force of attraction that indicates the role of a given simplex in the system. Here, simplex A has a negative t-force of 3, meaning that information or value is moving away from the simplex. On the other hand, simplex D has an attractive t-force of 10, meaning D is a possible sink or destination for information flows.

This opens up new possibilities in which we can use the q-analysis system to compute the dynamics of the system (complex). However, the dynamics of flow are difficult to measure by focusing only on the traffic values, as they do not tell us anything about where the information is flowing, or how it is being obstructed in the system. Moreover, one of the benefits of systems modeling is the ability to experiment with different policies and to assess who those policies affect and effected by the structure of the system. Up to this point we have focus primarily on describing the system using algebraic methods. In order to study the dynamics of flow or change, and the impact of traffic on the complex and vice versa, we need to introduce another new concept, 'q-Transmission.'

## Computing Transmission
In this notebook, I adopted Johnson's formal specification of the q-Transmission property. Q-transmission defines the relationship between the complex and the 'movement' or change in traffic as t-forces as indicated above. However, to this point we have focused primarily on system level properties with the exception of eccentricity measures. In Johnson's q-transmission theory, we focus on defining transmission fronts, the shared faces that allow movement of something from one simplex to another. Up to this point we have focused on computing q-connectivity as a general property of the system, but q-transmission relies on computing the q-near components, where a pair of simplicies share one or more vertices. While a pair of simplicies might be q-connected, it is not guaranteed that they are also q-near. For example, in our q-component at q = 0, all simplicies are q-connected. Yet, this does not tell us about flow at a given point of time. Now, there is a whole discussion about clock time versus social time and so on, but for now I will save this for a later date.

In general, we will simply consider the movement of information from one simplex to another given a shared vertex. This means traffic at 0-q can move directly from the point of origin (simplex) to another simplex sharing that vertex at t+1, where t = 0. This shared vertex represents the transmission 'front.' The traffic that is now present at a given simplex (t = 1) can move to other simplicies that are q-near. This movement can continue through q-near simplicies, and eventually the entire q-connected complex. The function below provides a starting point for computing the q-transmission values for any pair of simplicies.

```python
f = computeQNear(QMatrix)
f
```

The output above highlights the dimension and the q-near simplicies at that dimension. For example, at the dimension 0.0 A is q-near to 'D', 'B', 'C', 'E', 'H', 'I', 'L', meaning that A shares at least 1 vertex with each of these simplicies. This is also means that information or traffic defined by 0-dimension can exist on any one of these simplicies where t=1. Additionally, this indicates that the number of transmission fronts of A at 0-dimension are 2 or greater, where A = F_sub_0, 'D', 'B', 'C', 'E', 'H', 'I', 'L' = F_sub_1, and 'F', 'G', 'J', 'K' = F_sub_2 or greater.
The value of -1 in the dimension key of the dictionary represent simplicies (F', 'G', 'J', 'K') that do not share a common vertex with the simplex under consideration. However, as indicated previously, these simplicies still may be q-connected and our goal of computing the transmission fronts for each simplex begins by traversing the q-near simplicies until all q-near simplicies have been exhausted.

One simple way for computing the transmission fronts from the q-near simplicies is to consider all sets of simplicies and their q-near relations to determine if elements in -1 of A are in 0 of any other simplex except themselves. If so, then the simplex in -1 of A that is also 0 or more of another simplex, and that is also in 0 of A results in a transmission number of F = 2. Similarly, there are situations where a simplex might be many fronts away from another simplex and still be q-connected. Such a hypothetical system might start at A and move from B to C to D to reach L. In this case pair of simplicies A and L have transmission front number of F = 4.

Of course, this idea of transmission can be further defined by returning our focus to the conjugate complex of the incidence matrix. By focusing on the vertices, we can compute the exactly the links between each simplex and model the flows from one simplex to another based on a specific vertex or set of vertices. Johnson describes this process in the context of modeling traffic dynamics in transportation network where roads represent vertices links between junctions on-ramps or roundabouts that represent the nodes or simplicies. To model the flow of traffic, Johnson focused on measuring movement and identifying choke points in the system, the nodes. Here traffic can accumulate leading to roadway congestion, or flow freely indicating open roadways. It is important to note this example as Johnson explicitly models the links as fundamental properties of the system, but nearly anything could be described in this way where we would want to measure flows both locally and globally over the complex.
